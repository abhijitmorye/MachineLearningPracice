# Linear Regression with Single Variable

```python
df = pd.read_csv('ex1data1.txt', sep=",", header=None)
```

```python
df.columns = ['populations', 'profits']
```

```python
df
```

```python
ax = sns.scatterplot(x='populations', y='profits', data=df)
```

```python
# setting all data

m = df.shape[0] # number of training examples in our dataset
X = np.hstack((np.ones((m, 1)),df.populations.values.reshape(-1,1))) # m X (n+1) dimensinal vector
# y1 = df.profits.values.reshape(-1,1)
y2 = np.array(df.profits.values).reshape(-1,1) # m X 1 column vector
theta = np.zeros(shape=(X.shape[1], 1)) # (n+1) X 1 vector since theta0 and theta1
iterations = 1500
alpha = 0.01
```

```python
def compute_cost(X, y, theta):
    m = X.shape[0]
    h = X.dot(theta)
    J = (1/(2*m)) * (np.sum((h-y)**2))
    return J
```

```python
cost = compute_cost(X, y2, theta)
cost
```

```python
# for theta [[-1], [2]]
cost1 = compute_cost(X, y2, [[-1], [2]])
cost1
```

```python
# gradient descent


def gradient_descent(X, y, theta, alpha, num_iter):
    m = y.shape[0]
    j_history = np.zeros(shape=(num_iter, 1))
    for i in range(0, num_iter):
        h = X.dot(theta)
        diff_h = h -y
        delta = (1/m) * (diff_h.T.dot(X))
        theta = theta - (alpha * delta.T)
        print(f"Cost at iteration {i} is {compute_cost(X, y, theta)}")
        j_history[i] = compute_cost(X, y, theta)
    return theta, compute_cost
```

```python
theta, costs = gradient_descent(X, y2, theta, alpha, iterations)
theta
```

```python
# plotting linear regression line


ax = sns.scatterplot(x='populations', y='profits', data=df)
plt.plot(X[:,1], X.dot(theta), color='r')
```

```python
# Prediction

y_pred = np.array([1, 3.5]).dot(theta)
f'For population = 35,000, we predict a profit of {y_pred[0]*10000}'
```

```python
y_pred = np.array([1, 7]).dot(theta)
f'For population = 70,000, we predict a profit of {y_pred[0]*10000}'
```

# Linear Regression with Multiple Variables

```python
df2 = pd.read_csv('ex1data2.txt', sep=',', header=None)
df2.columns = ['house_size', 'bedrooms', 'house_price']
df2.describe().T
```

<p>
    By looking at the values, note that house sizes are about 1000 times the number of bedrooms. When features differ by orders of magnitude, first performing feature scaling can make gradient descent converge much more quickly.

We can speed up gradient descent by having each of our input values in roughly the same range, ideally  i.e <b><i> -1.0 < x < 1.0 or -0.5 < x < 0.5 </i></b>
</p>

<ul>
    <li>Feature scaling: involves dividing the input values by the range (i.e. the maximum value minus the minimum value) of the input variable </li>
    <li> Mean normalization: involves subtracting the average value for an input variable from the values for that input variable </li>
</ul>

<p>
    <b><i> x-u/s </i></b>
    
    <ul>
        <li> x = data value </li>
        <li> u = data mean value </li>
        <li> s = data std value </li>
    </ul>
</p>

```python
def feature_normalized(X, mean=np.zeros(1), std=np.zeros(1)):
    
    X = np.array(X)
    if len(mean.shape)== 1 and len(std.shape) == 1:
        mean = np.mean(X, axis=0)
        std = np.std(X, axis=0, ddof=1)
    X_norm = (X-mean)/std
    return X_norm, mean, std
```

```python
X_norm, mean , std = feature_normalized(df2[['house_size', 'bedrooms']])
df2['house_size_normalized'] = X_norm[:,0]
df2['bedrooms_normalized'] = X_norm[:,1]
df2[['house_size_normalized', 'bedrooms_normalized']].describe().T
```

```python
# Calculating cost
def compute_cost_mulyiple_varibles(X, y, theta):    
    m = y.shape[0]
    h = X.dot(theta)
    j = (1/(2*m)) * ((h-y).T.dot(h-y))
    return j
```

```python
def gradient_descent_mult_var(X, y, theta, alpha, num_iter):
    m = y.shape[0]
    j_history = np.zeros(shape=(num_iter, 1))
    
    for i in range(0, num_iter):
        h = X.dot(theta)
        diff_h = h - y
        
        delta = (1/m)* (diff_h.T.dot(X))
        theta = theta - (alpha*delta.T)
        j_history[i] = compute_cost_mulyiple_varibles(X, y, theta)
        print(f"Cost at iteration {i} is {compute_cost_mulyiple_varibles(X, y, theta)}")
    print("*"*100)
    return theta, j_history
```

```python
# setting values
m = df2.shape[0]
X = np.hstack((np.ones((m, 1)), X_norm))
y = np.array(df2.house_price.values).reshape(-1,1)
theta = np.zeros(shape=(X.shape[1], 1))
```

```python
# initial cost
cost_mul_var = compute_cost_mulyiple_varibles(X, y, theta)
cost_mul_var
```

```python
alpha = [0.3, 0.1, 0.03, 0.01]
color = ['b', 'r', 'g', 'c']
num_iters = 50

for i in range(len(alpha)):
    print(f"For theta value {alpha[i]}")
    theta = np.zeros(shape=(X.shape[1], 1))
    theta, j_history = gradient_descent_mult_var(X, y, theta, alpha[i], num_iters)
    plt.plot(range(len(j_history)), j_history, color=color[i], label=f"Aplha {alpha[i]}")
```

```python
# using theta value 0.3

theta, gradient_descent_mul_var = gradient_descent_mult_var(X, y, theta, 0.3, 250)
theta
```

```python
# prediction


sqft = (1650 - mean[0])/std[0]
bedrooms = (3 - mean[1])/std[1]
y_pred = theta[0] + theta[1]*sqft + theta[2]*bedrooms
f'Price of a house with 1650 square feet and 3 bedrooms: {y_pred[0]}$'
```

# Normal Equations

<p>
    A closed-form solution to find  without iteration.
            theta = inverse(X.T.dot(X)).dot(X.T.dot(y))
</p>      

```python
def normal_equ(X,y):
    inv = np.linalg.pinv(X.T.dot(X))
    theta = inv.dot(X.T).dot(y)
    return theta
```

```python
Xe= np.hstack((np.ones((X.shape[0], 1)), df2[['house_size', 'bedrooms']].values))
ye = np.array(df2.house_price.values).reshape(-1,1)
theta_e = normal_equ(Xe, ye)
theta_e
```

```python
y_pred = theta_e[0] + theta_e[1]*1650 + theta_e[2]*3
f'Price of a house with 1650 square feet and 3 bedrooms: {y_pred[0]}$'
```
